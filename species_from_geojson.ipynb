```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
data = gpd.read_file("data/points_fish_cro/points_fish_cro_aqua_wgs84.geo.json")
```

```python
data
```

```python
labels = []
for i in data:
    if "species" in i:
        labels.append(i)
```

```python
species = []
for i in labels:
    species.append(data[i].unique())
```

```python
species = np.concatenate(species)
```

```python
sp = []
for i in species:
    if i != None:
        sp.append(i)
```

```python
latin = []
for i in sp:
    if "Agryrosomus" not in i: 
        if "Mytilus" not in i:
            if  "Ostrea" not in i:
                if "Dicentrachus labrax" in i:
                    latin.append("Dicentrarchus labrax")
                elif "Diplodus sargus sargus" in i:
                    latin.append("Diplodus sargus")
                else:
                    latin.append(i[i.find("(")+1:i.find(")")].replace("  ", " "))
```

```python
latin = np.unique(latin)
```

```python
latin
```

```python
df = pd.DataFrame({"species":latin})
```

```python
df.to_csv("data/species_list_hrv.csv")
```
